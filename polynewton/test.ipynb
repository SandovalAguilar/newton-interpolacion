```python
import numpy as np
import matplotlib.pyplot as plt
import polynewton_method as pm

%matplotlib inline
```

```python
# Funcion para generar el grafico
def draw_plot(newton_poly, name):
    plt.figure(figsize = (6, 4))
    plt.plot(newton_poly.x_vector, newton_poly.y_vector, 'bo')
    plt.plot(newton_poly.x_values, newton_poly.evaluated_polynomial)
    plt.title(name)
    plt.show()
```

#### Casos de prueba

```python
x_vector_1 = np.array([0, -2, 3])
y_vector_1 = np.array([-6, -8, 3])
# Object creation
test_1 = pm.PolyNewton(x_vector_1, y_vector_1, 3)
print("Diferencias divididas:")
print(test_1.calculate_coefficients())
print("\nValores de X:")
print(np.round(test_1.x_values, 1))
print("\nPolinomio evaluado:")
print(test_1.evaluate_interpolation())
```
